## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
imdb['Gross'].sum()
```

```python
# Average "No_of_Votes" of all movies
imdb['No_of_Votes'].mean()
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
imdb.groupby('Decade').agg({'Series_Title': 'count'}).sort_values(by='Series_Title', ascending=False)
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
imdb.groupby('Director').agg({'Series_Title': 'count'}).sort_values(by='Series_Title', ascending=False).head(10)
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
imdb.groupby('Director').agg({'Series_Title':'count', 'IMDB_Rating':'mean', 'Gross':'sum'}).sort_values(by='Gross', ascending=False).head(10).reset_index()
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count
imdb.groupby(['Decade', 'Director']).agg({'Series_Title':'count', 'IMDB_Rating':'mean'}).sort_values(by='Series_Title', ascending=False).reset_index()
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking
recent_movies=imdb.loc[imdb['Released_Year']>=2010]
recent_movies['Ranking']=recent_movies.groupby('Released_Year').cumcount()+1
recent_movies=recent_movies.sort_values(by=['IMDB_Rating'], ascending=[False])
Top_3_recent_movies=recent_movies[['Released_Year','Ranking','Series_Title','IMDB_Rating']].head(3)
Top_3_recent_movies.sort_values(by=['Released_Year','Ranking'], ascending=[False,True])
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.
directors_decade=imdb.groupby(['Director', 'Decade']).agg({'Series_Title':'count'}).reset_index()
directors_decade.loc[directors_decade['Series_Title']>=5]
```
