## Question 4. Adding new columns

Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers.

For example, 759 means 7:59 and 801 means 8:01. Their difference is not 42 but 2 minutes.

```python
# Convert them to a more convenient representation of number of minutes since midnight (0).
flights['dep_time_min'] = flights['dep_time'] // 100 * 60 + flights['dep_time'] % 100

flights['sched_dep_time_min'] = flights['sched_dep_time'] // 100 * 60 + flights['sched_dep_time'] % 100

display(flights[['dep_time', 'dep_time_min', 'sched_dep_time', 'sched_dep_time_min']].head())

```

```python
# Create a new column of arr_time - dep_time.
flights['arr_time_min'] = flights['arr_time'] // 100 * 60 + flights['arr_time'] % 100
flights['calculated_flight_duration'] = flights['arr_time_min'] - flights['dep_time_min']

# Compare this column with air_time.
display(flights[['calculated_flight_duration', 'air_time']].head())

display((flights['calculated_flight_duration'] - flights['air_time']).head())
```

## Question 5. Mixing things together

The following questions may require multiple operations above.

```python
# Find the 20 most delayed flights.
# Display the following: year,month,day,carrier,flight,dep_delay,arr_delay,carrier
# How do you want to handle ties?

# Sort by 'dep_delay' in descending order and select the top 20.
# Using nlargest to handle ties by including all rows with the same largest values.
most_delayed_flights = flights.nlargest(20, 'dep_delay', keep='all')

# Display the requested columns
display(most_delayed_flights[['year', 'month', 'day', 'carrier', 'flight', 'dep_delay', 'arr_delay']])
```

```python
# Sort all AA flights to find the top 10 fastest (highest speed) flights.
# Display the following: year,month,day,carrier,flight,orig,dest,distance,air_time,speed (miles per hour)
AA_flights = flights[flights['carrier'] == 'AA']
AA_flights['speed'] = AA_flights['distance'] / (AA_flights['air_time'] / 60)
display(AA_flights.sort_values(by='speed', ascending=False).head(10))
```

```python
# Find all flights that satisfy the following:
# - From John F. Kennedy Airpot (JFK) or Newark Aiport (EWR) to Seattle-Tacoma Airport (SEA)
# - Carrier is UA, AA, or DL.
# - Dates from 4/1/2013 (inclusive) to 4/3/2013 (inclusive)
# - Scheduled arrival time is before noon.
# - Display the following: year,month,day,carrier,flight,origin,dest,sched_dep_time,sched_arr_time
# - Sort by year, month, day, sched_arr_time

all_flights = flights[(flights['origin'].isin(['JFK', 'EWR'])) & (flights['dest'] == 'SEA') & (flights['carrier'].isin(['UA', 'AA', 'DL'])) & (flights['year'] == 2013) & (flights['month'] == 4) & (flights['day'] >= 1) & (flights['day'] <= 3) & (flights['sched_arr_time'] < 1200)]
display(all_flights.sort_values(by=['year', 'month', 'day', 'sched_arr_time']))
all_flights.sort_values(by=['year', 'month', 'day', 'sched_arr_time'])
```
